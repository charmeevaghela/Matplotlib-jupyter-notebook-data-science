```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6]
y = [23,90,1,45,33,20]
#plt.plot(x,y,color="red",linewidth=4,linestyle="dashdot")   # color and linewidth, linestyle can be changed
```

## x-label y-label and title can also be set
## sometimes you need to restart the kernel and run the code again if code is not working 

```python
plt.xlabel('index')
plt.ylabel('element')
plt.title('Index Element')
plt.plot(x,y)
```

# below 3rd argument is for format string r=> red color , D => marker diamong shape visit the documentation for more on this 
# there are different Markers like + , D, * etc. , -- => is to join markers , color, marker and joining can be specified in any order

```python
plt.plot(x,y,'rD--') 
```

```python
plt.plot(x,y,'--g+',markersize=20) # to change the marker size
```

```python
plt.plot(x,y,alpha=0.6) # alpha is for transperancy of the plot
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

## ploting more than 1 charts in one chart

## that small box shown in the below plot is called legend, plt.legend() is to be set to show it. and each plot should have a label to be shown in the legend , if label is not given but plt.legend() is set then legend won't be shown

```python
plt.xlabel("week day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t,label="max") 
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc="upper right", fontsize="large") # location of the legend can also be provided , 
# plt.legend(loc="best") auto locate according to space available

# fontsize can be small, medium, large
```

## grid can be set using plt.grid() 

```python
plt.xlabel("week day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t,label="max") 
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc="best")
plt.grid()
```
