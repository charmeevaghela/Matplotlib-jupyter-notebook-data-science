```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
%%HTML
<h1 align='center' style="color:red;"> Bar Chart </h1>
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)  # 0 => googl , 1 => amzn, 2=>msft , 3=> fb
plt.bar(xpos,revenue)
```

```python
profit = [40,2,34,12]
```

```python
plt.bar(xpos,revenue,label="revenue")
plt.bar(xpos,profit,label="profit")
plt.legend()
```

## now i want this side by side

```python
plt.bar(xpos-0.2,revenue,label="revenue")
plt.bar(xpos+0.2,profit,label="profit")
plt.legend()
```

## width added

```python
plt.bar(xpos-0.2,revenue,width=0.4,label="revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

## horizontal bar chart

```python
plt.yticks(xpos,company)
plt.barh(xpos-0.2,revenue,height=0.4,label="revenue")
plt.barh(xpos+0.2,profit,height=0.4,label="profit")
plt.legend()
```

### For histogram we only need a one dimensional array

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar_men)
# below the second array is called bins ( or say containers), we can create our own bins 
```

```python
plt.hist(blood_sugar_men,bins=[80,100,125,150],rwidth=0.95) # this means 80 to 100, 100 to 125, 125 to 150
```

### To plot for two arrays 

```python
plt.xlabel('sugar level')
plt.ylabel('no of patients')
plt.title("Blood sugar level analysis")
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['orange','blue'],label=['men','women'])
plt.legend()
```

```python
%%HTML
<h1 align='center' style="color:red;"> Pie Chart </h1>
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/Internet Bill","Car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
 
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.4)
plt.show() # hides additional info 
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.4,autopct="%0.2f%%") # 2 digitd after decimal point
plt.show() # hides additional info 
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.4,autopct="%0.2f%%", shadow=True,explode=[0,0,0,0.5,0.2]) # 2 digitd after decimal point
plt.show() # hides additional info 
```

### by default start angle is 0, to change it use startangle attribute

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.4,startangle=30,autopct="%0.2f%%",shadow=True,explode=[0,0,0,0.5,0.2]) # 2 digitd after decimal point
plt.show() # hides additional info 
```
